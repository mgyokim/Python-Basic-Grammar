# 딕셔너리 만들 때 주의할 사항

```python
# Key 는 유니크값이다. 
# 다음 예에서 볼 수 있듯이 동일한 Key 가 2개 존재할 경우, 1: 'a' 쌍이 무시된다.
a = {1: 'a', 1: 'b'}
print(a)
```

```python
# Key 에는 변하는 값은 사용하지 못한다. 
# Key 에 리스트는 사용할 수 없고, 튜플은 사용할 수 있다. mutable 인지, immutable 값인지가 중요.
a = {[1, 2]: 'hi'}
```

```python
a = {(1, 2, 3): 'hi'}
print(a)
```

# 딕셔너리 관련 함수

# Key 리스트 만들기 - keys

```python
# a.keys() 는 딕셔너리 a의 key 만을 모아 dict_keys 객체를 반환한다.
a = {'name': 'pey', 'phone': '010-9999-1234', 'birth': '1118'}
print(a.keys())
print(a.values())

# for 문에서 바로 사용할 수도 있고,
for i in a.keys():
    print(i)
# list() 함수로 리스트로 변환할 수 도 있다.
print(list(a.keys()))
```

# Value 리스트 만들기 - values

```python
print(a.values())
```

# Key, Value 쌍 얻기 - items

```python
print(a.items())
```

# Key 로 Value 얻기 - get

```python
a = {'name': 'pey', 'phone': '010-9999-1234', 'birth': '1118'}
print(a.get('name'))

print(a.get('phone'))
```

# a['nokey'] 와 a.get('nokey') 차이는 키가 없는 것에 대한 응답의 차이!

```python
# a['nokey']처럼 딕셔너리에 존재하지 않는 키로 값을 가져오려고 할 경우, 
# a['nokey'] 방식은 오류를 발생시키고 
# a.get('nokey') 방식은 None을 반환한다는 차이
```

```python
print(a.['nokey'])
```

```python
print(a.get('nokey'))
```

```python
# Key 가 반드시 존재한다고 확실할 수 있을 때는 a[key] 방식을, 
# Key 가 없을 수도 있을 때는, a.get(key) 방식을 사용하면 된다.
```

# 딕셔너리 안에 찾으려는 Key 가 없을 경우, 미리 정해준 디폴트 반환하기.

```python
print(a.get('nokey', '정보없음'))
```

# 해당 Key가 딕셔너리 안에 있는지 조사하기 - in

```python
a = {'name': 'pey', 'phone': '010-999-1234', 'birth': '1118'}
print('name' in a)

print('email' in a)
```

# Key로 Value꺼내기 - pop

```python
a = {'name': 'pey', 'phone': '010-999-1234', 'birth': '1118'}
phone = a.pop('phone')
print(phone)
print(a)
```

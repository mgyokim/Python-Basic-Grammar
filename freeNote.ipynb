# 실수형

```python
# 실수형
a = 1.2
print("🚀 ~  ~ a: ", a);
a = -3.45
print("🚀 ~  ~ a: ", a);

a = 4.24E10
print("🚀 ~  ~ a: ", a);

a = 0o177 #0o177 = 1 * 8^2 + 7 * 8^1 + 7 * 8^0 = 127
print("🚀 ~  ~ a: ", a); # 8진수 octal

a = 0x8ff
print("🚀 ~  ~ a: ", a);  # 16진수 hexadecial (8 * 16^2) + (15 * 16) + 15 = 2303

b = 0xABC
print("🚀 ~  ~ b: ", b); # (10 * 16^ 2) + (11 * 16) + 12 = 2748
```

# 문자열 자료형

```python
# 큰따옴표로 양쪽 둘러싸기
print("Hello, world!")
# 작은따옴표로 양쪽 둘러싸기
print('Hello, world!')
# 작은따옴표 3개를 연속으로 써서 양쪽 둘러싸기
print('''Hello, world!''')
# 큰따옴표 3개를 연속으로 써서 양쪽 둘러싸기
print("""Hello, world!""")
# 문자열에 작은따옴표 포함하기
print("Python's favorite food is perl")
# 문자열에 큰따옴표 포함하기
print('Python"s favorite food is perl')
# 역슬래시를 사용해서 작은따옴표와 큰따옴표를 문자열에 포함하기
print('Python\'s favorite food is perl')
print("\"Python is very easy.\" he says")
```

# 정렬과 공백

```python
a = "%10s" % "hi" # 전체 길이가 1-0개인 문자열 공간에서 대입되는 값을 오른쪽으로 정렬하고 그 앞의 나머지는 공백으로 남겨 두라는 의미.
print(a)
a = "%-10sjane" % "hi" # 전체 길이가 1-0개인 문자열 공간에서 대입되는 값을 왼쪽으로 정렬하고 그 앞의 나머지는 공백으로 남겨 두라는 의미. jane 은 제외하고 길이 10개보장임.
print(a)
```

# 문자열이 알파벳으로만 구성되어 있는지 확인하기 - isalpha

# 정수형

```python
# 정수형
a = 123
print("🚀 ~  ~ a: ", a);

a = -178
print("🚀 ~  ~ a: ", a);
a = 0
print("🚀 ~  ~ a: ", a);
```

# 실수형

```python
# 실수형
a = 1.2
print("🚀 ~  ~ a: ", a);
a = -3.45
print("🚀 ~  ~ a: ", a);

a = 4.24E10
print("🚀 ~  ~ a: ", a);

a = 0o177 #0o177 = 1 * 8^2 + 7 * 8^1 + 7 * 8^0 = 127
print("🚀 ~  ~ a: ", a); # 8진수 octal

a = 0x8ff
print("🚀 ~  ~ a: ", a);  # 16진수 hexadecial (8 * 16^2) + (15 * 16) + 15 = 2303

b = 0xABC
print("🚀 ~  ~ b: ", b); # (10 * 16^ 2) + (11 * 16) + 12 = 2748
```

# 사칙연산

```python
# 사칙 연산
a = 3
b = 4
a + b
print("🚀 ~  ~ a + b: ", a + b);
a - b
print("🚀 ~  ~ a - b: ", a - b);
a * b
print("🚀 ~  ~ a * b: ", a * b);
a / b
print("🚀 ~  ~ a / b: ", a / b);
```

# x의 제곱을 나타내는 ** 연산자

```python
a = 3
b = 4
print("🚀 ~  ~ a ** b: ", a ** b);
```

# 나눗셈 후 몫을 반환하는 // 연산자

```python
print("🚀 ~  ~ 7 / 4: ", 7 / 4);
print("🚀 ~  ~ 7 // 4: ", 7 // 4);
```

# 복합 연산자

```python
a = 1
a = a + 1
print("🚀 ~  ~ a: ", a);
a += 1
print("🚀 ~  ~ a: ", a);

a -= 1
print("🚀 ~  ~ a: ", a);
```

# 문자열 자료형

```python
# 큰따옴표로 양쪽 둘러싸기
print("Hello, world!")
# 작은따옴표로 양쪽 둘러싸기
print('Hello, world!')
# 작은따옴표 3개를 연속으로 써서 양쪽 둘러싸기
print('''Hello, world!''')
# 큰따옴표 3개를 연속으로 써서 양쪽 둘러싸기
print("""Hello, world!""")
# 문자열에 작은따옴표 포함하기
print("Python's favorite food is perl")
# 문자열에 큰따옴표 포함하기
print('Python"s favorite food is perl')
# 역슬래시를 사용해서 작은따옴표와 큰따옴표를 문자열에 포함하기
print('Python\'s favorite food is perl')
print("\"Python is very easy.\" he says")
```

# 이스케이프 코드 예제

```python
# \r	캐리지 리턴(줄 바꿈 문자, 커서를 현재 줄의 가장 앞으로 이동)
import time

for i in range(1, 101):
    print(f"\r진행률: {i}%", end="")
    time.sleep(0.01)
```

# 문자열 인덱싱과 슬라이싱


```python
a = 'Life is too short, You need Python'
print(a[3]) # e
print(a[0])
print(a[-0])
print(a[-1])
print(a[-2])
print( a[0] + a[1] + a[2] + a[3])
print(a[0:4])
print(a[0:3])
print(a[0:2])
print(a[0:1])
print(a[0:-2]) # Life is too short, You need Pyth
```

# 슬라이싱으로 문자열 나누기


```python
a = "20230331Rainy"
data = a[:8]
print("🚀 ~  ~ data: ", data);
weather = a[8:]
print("🚀 ~  ~ weather: ", weather);
year = a[:4]
print("🚀 ~  ~ year: ", year);
day = a[4:8]
print("🚀 ~  ~ day: ", day);
```
